```python
def calculate_intensity(compound):
    if compound >= 0:
        if compound > 0.65:
            return 4
        elif compound > 0.4:
            return 3
        elif compound > 0.3:
            return 2
        else:
            return 1
    elif compound < 0:
        if compound > -0.2:
            return 1
        elif compound > -0.3:
            return 2
        elif compound > -0.4:
            return 3
        elif compound > -0.8:
            return 4
        else:
            return 5
```

```python
# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create lists to store results
data = []

# Analyze each sentence
for sentence in sentences:
    # VADER analysis
    vader_scores = analyzer.polarity_scores(sentence)
    
    # TextBlob analysis
    # textblob_score = TextBlob(sentence).sentiment.polarity

    # Append results
    data.append([
        sentence,
        # vader_scores['pos'],
        # vader_scores['neu'],
        # vader_scores['neg'],
        calculate_intensity(vader_scores['compound']),
        # textblob_score
    ])

# Convert to DataFrame
df_vader = pd.DataFrame(data, columns=["Sentence",  "VADER_Compound"])
# sample_size = int(len(df_vader)*(0.934))
# df_new.loc[:sample_size,'Intensity']= df_vader.loc[:sample_size,'VADER_Compound']
# df_new.to_csv('./data/new_synth2.csv', index=False)
# Display DataFrame in a formatted table
print(tabulate(df_vader, headers="keys", tablefmt="fancy_grid", showindex=False))
```

```python
# check accuracy of the intensity values
correct = 0
for i in range(len(df)):
    if df.loc[i,'Intensity'] == df_vader.loc[i,'VADER_Compound']:
        correct += 1
print(f"Accuracy: {correct/len(df)}")
```
